```python
import pandas as pd
df=pd.read_csv("customer_shopping_behavior.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.isnull().sum())

```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower() # converting upper case columns into lower case and replacing spaces with underscore to make easy for analysis
df.columns=df.columns.str.replace(' ','_') 
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
df['age_Group'] = pd.qcut(df['age'],q=4,labels=['Young', 'Adult', 'Middle-Aged', 'Senior'])
```

```python
df.head()
```

```python
freq_map={
    "Fortnightly":14,
    "Weekly":7,
    "Annually":365,
    "Bi-Weekly":14,
    "Monthly":30,
    "Quarterly":90,
    "Every 3 Months":90
}
df["frequency_of_purchases"]=df["frequency_of_purchases"].map(freq_map)
```

```python
df.head(10)
```

```python
print((df['discount_applied']==df['promo_code_used']).all())
```

```python
df[['discount_applied','promo_code_used']]=df[['discount_applied','promo_code_used']].replace({'Yes':1,'NO':0})
```
